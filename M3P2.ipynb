```python
import pandas as pd
import numpy as np


dades = pd.read_csv('DelayedFlights.csv')
print(dades.keys())
dades = dades.sample(200000)
```

Hi ha molta informació respecte els retards en els vols. Per fer un estudi estadísitc senzill, anem a agafar les següents variables:
 - 'UniqueCarrier' perquè sembla ser l'operadora, juntament amb 'FlightNum' el número de vol.
 - 'ActualElapsedTime', 'CRSElapsedTime' per ser els temps, real i programat respectivament, de la durada del viatge. A més, assignem la columna 'Retard' que indica si l'avió es va retrasr o no.

```python
dades = dades[['UniqueCarrier', 'FlightNum', 'CRSElapsedTime', 'ActualElapsedTime']]
dades['Retard'] = ~(dades['CRSElapsedTime'] >= dades['ActualElapsedTime'])
print(dades)
```

```python
vols_operadores = {}
for item in dades['UniqueCarrier']:
    if item not in vols_operadores.keys():
        vols_operadores[item] = 1
    else:
        vols_operadores[item] += 1

dades_agrupades = {}
for key in vols_operadores.keys():
    dades_aux = dades[dades['UniqueCarrier'] == key]
    dades_agrupades[key] = [dades_aux[dades_aux['Retard'] == True].shape[0]/vols_operadores[key], dades_aux[dades_aux['Retard'] == False].shape[0]/vols_operadores[key]]

dades_agrupades = pd.DataFrame.from_dict(dades_agrupades)*100
dades_agrupades = dades_agrupades.rename(index={0: 'OnTime (%)', 1: 'Delayed (%)'})
dades_agrupades = dades_agrupades.transpose()
print('L\'operadora amb més retards és la següent:')
print(dades_agrupades[dades_agrupades['Delayed (%)'] >= dades_agrupades['Delayed (%)'].max()])
```

Cosiderarem els vols més llargs com el vols amb una durada de fins al 85% del vol més llarg de tots.

```python
print('Els vols més llargs són:')
dades[dades['ActualElapsedTime'] >= 0.85 * dades['ActualElapsedTime'].max()]
```
